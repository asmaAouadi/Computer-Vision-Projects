```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
readimage = cv2.imread('im1.jpg')


plt.figure(figsize=(6,4))
imgFirstGaussian = cv2.GaussianBlur(readimage , (3,3) , 4 , None , 0)
for _ in range(5):
    imgGaussianBlur = cv2.GaussianBlur(imgFirstGaussian, (3,3),4,None, 0)  
    imgFirstGaussian = imgGaussianBlur
    plt.imshow(imgGaussianBlur)
    plt.axis('off')
    plt.show()
     
```
